```python
teamrun=df.groupby("Team")["Runs"].sum()
df["Contribution%"]=df["Runs"]/teamrun[df["Team"]].values *100
df.head()
```

```python
df["boundary_balls"]= df["4s"] + df["6s"]
df["runs_from_boundaries"] = df["4s"] *4+ df["6s"]*6
df["runs_from_non_boundaries"] = df["Runs"] - df["runs_from_boundaries"]
df["non_boundary_balls"] = df["BF"] - df["boundary_balls"]
df["dot_balls_est"] = df["non_boundary_balls"] - df["runs_from_non_boundaries"]
df.head()
```

```python
df["ImpactScore"] = (
    (df["Runs"] / df["Runs"].max()) * 0.4   # 40% weight
    + (df["SR"] / df["SR"].max()) * 0.35    # 35% weight
    + (df["Contribution%"] / 100) * 0.25    # 25% weight
) * 100
```
